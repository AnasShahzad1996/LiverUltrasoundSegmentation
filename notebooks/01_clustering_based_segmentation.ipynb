```python
# Read the image
import cv2
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Read the image
image_path = '/home/anas/Desktop/code/practikum/our_code/datasets/patient5/Patient-05-felixH-01131.png'
image = cv2.imread(image_path)

# Reshape the image to a 2D array of pixels
pixels = image.reshape(-1, 3)

# Perform K-means clustering
k = 6  # Number of clusters
kmeans = KMeans(n_clusters=k, random_state=0)
kmeans.fit(pixels)

# Get the cluster labels for each pixel
labels = kmeans.labels_

# Reshape the labels back to the original image shape
segmented = labels.reshape(image.shape[:2])

# Define RGB colors for each segment (replace with your desired colors)
colors = [
    [255, 0, 0],   # Red
    [0, 255, 0],   # Green
    [0, 0, 255],   # Blue
    [255, 255, 0], # Yellow
    [255, 0, 255], # Magenta
    [0, 255, 255]  # Cyan
]

# Generate the segmented image with colors
segmented_colored = np.zeros_like(image)
for i in range(k):
    segmented_colored[segmented == i] = colors[i]

# Display the original image and the segmented image
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

axs[0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
axs[0].set_title('Original Image')
axs[0].axis('off')

axs[1].imshow(cv2.cvtColor(segmented_colored, cv2.COLOR_BGR2RGB))
axs[1].set_title('Segmented Image')
axs[1].axis('off')

plt.show()
```
